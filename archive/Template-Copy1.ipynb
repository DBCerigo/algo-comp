```python
test_cases = (
    ((5, [3,4,4,6,1,4,4]), [3, 2, 2, 4, 2]),
    ((5, [3]), [0, 0, 1, 0, 0]),
    ((5, [6]), [0, 0, 0, 0, 0]),
    ((5, [3,6]), [1, 1, 1, 1, 1]),
    ((1, [3,6]), [0,]),
    ((1, [1]), [1]),
    ((1000, [1001]*1000),[0]*1000)
    )
```

```python
def solution(N, A):
    result = [0]*N
    for opp in A:
        if opp > N:
            # make max
            result = [max(result)]*N
        else:
            result[opp-1] += 1
    return result
```
