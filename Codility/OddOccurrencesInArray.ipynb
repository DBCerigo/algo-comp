```python
test_cases = [
    ([1,2,1], 2),
    ([1,], 1),
    ([1,1,1,1,5], 5),
    ([2] + ([10000]*1000), 2)
]
```

```python
from collections import Counter
def solution(A):
    count_dict = Counter(A)
    for k, v in count_dict.items():
        if v % 2 == 1:
            return k
```

```python
for case in test_cases:
    try:
        print(solution(case[0]))
        print(case)
    except Exception as e:
        print(e)
```
